```python
df_work = df[["Gender", "Date_Diff", "Age", "Scholarship", "Hipertension", "Alcoholism", "Handcap", "SMS_received", "No-show"]]
```

```python
df_work
```

```python
##Replace M with 1 and F with 0 in Gender
df_work["Gender"].replace({"M": 1}, inplace=True)
df_work["Gender"].replace({"F": 0}, inplace=True)
```

```python
df_work
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = df_work.drop(columns=["No-show"])
y = df_work["No-show"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
```

```python
y_pred = dt_model.predict(X_test)
y_pred
```

```python
score = accuracy_score(y_test, y_pred)
score
```
